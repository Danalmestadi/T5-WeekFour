# Import libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
from sklearn.metrics import confusion_matrix, accuracy_score
```

# Load Data and EDA

```python
df = pd.read_csv('Churn_Modelling_Exercise.csv')
df.head()
```

```python
df.columns
```

## Preprocessing

### Drop unnecessary columns

```python
df.drop(['customerID', 'gender' ], axis=1, inplace=True)
df.head()
```

### One-hot-encoder and Label-encoder

```python
label_encoder = LabelEncoder()
df['Churn'] = label_encoder.fit_transform(df['Churn'])
```

```python
df_cat = df.select_dtypes(include='object')
df_num = df.select_dtypes(exclude='object')
df_cat.head()
OneHotEncoder = OneHotEncoder()
df_cat_encoded = pd.DataFrame(OneHotEncoder.fit_transform(df_cat).toarray())
```

```python
df=pd.concat([df_num,df_cat_encoded],axis=1)
df.head()
```

## Selecting Features && Train Test Split

### Selecting Features

```python
x=df.drop('Churn',axis=1)
y=df['Churn']
```

### Train Test Split

```python
train_test_split = train_test_split(x, y, test_size=0.2, random_state=42)
x_train, x_test, y_train, y_test = train_test_split
```

## Modeling

### Creating the sequential model

```python
model=Sequential()
```

### Adding the input layer

```python
model.add(Dense(units=6, input_dim=x_train.shape[1], kernel_initializer='uniform',activation='relu'))
```

### Adding the hidden layers

```python
model.add(Dense(units=6, activation='relu'))
model.add(Dense(units=6, activation='relu'))
```

### Adding the output layer

```python
model.add(Dense(units=1, activation='sigmoid'))
```

### Compiling the model

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

### Display model summary

```python
model.summary()
```

### Fitting the model

```python
model.fit(x_train, y_train, batch_size=10, epochs=100, validation_split=0.2)
```

## Evaluating the model

```python
model.evaluate(x_test, y_test)
```
